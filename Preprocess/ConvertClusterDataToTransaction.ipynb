```python
import pandas as pd
import os
```

```python
data = pd.read_csv('..\\DataMining\\src\\data\\NewDataAfterCluster.csv')
data
```

```python
clusters = data['cluster'].unique()
n_cluster = len(clusters)
clusters
```

```python
cluster_data_list = [data[data['cluster']==i] for i in clusters]
```

```python
pivot_tab_list = [pd.pivot_table(data=i, index='InvoiceNo', columns='StockCode',values='Quantity', aggfunc='sum',fill_value=0) for i in cluster_data_list]
pivot_tab_list
```

```python
transaction_tab_list = [i.applymap(lambda x: 1 if x > 0 else 0) for i in pivot_tab_list]
```

```python
transaction_tab_list
```

```python
for i in transaction_tab_list:
    i.index.name = None
    i.columns.name = None
```

```python
transaction_tab_list
```

```python
for i in range(len(transaction_tab_list)):
    transaction_tab_list[i].to_csv("..\\DataMining\\src\\data\\cluster_"+str(i+1)+"_transaction_data.csv", index=None)
```
